```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from scipy.misc import derivative
x = symbols('x')
```

### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



```python
def func(x):
    return x**5 + 4*sin(2*x) + cos(3*x+3)

der1 = derivative(func, 1.0, dx=1e-6)
der2 = derivative(func, 1.0, dx=1e-6, n=2)
print('Производная в точке x=1:', round(der1,3),'\nВторая производная в точке x=1:', round(der2,3))
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
def f(x):
    return x ** 2 + 1

x = np.linspace(-5,5)
der1 = [derivative(f, x0, dx=1e-6) for x0 in x]
der2 = [derivative(f, x0, dx=1e-6, n=2) for x0 in x]
    
plt.plot(x, f(x), label='function')
plt.plot(x, der1, label='derivative')
plt.plot(x, der2, label='second derivative')
plt.legend()
plt.show()
```
